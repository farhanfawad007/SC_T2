```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset directly from an online public source URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Display data information to check structure
print("--- Dataset Structure ---")
df.info()
```

```python
# 1. Fill missing values in 'Age' with the middle (median) age of passengers
df['Age'].fillna(df['Age'].median(), inplace=True)

# 2. Fill missing values in 'Embarked' with the most frequent port ('S')
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# 3. Remove the 'Cabin' column entirely since it is mostly blank rows
df.drop(columns=['Cabin'], inplace=True)

# 4. Confirm there are zero missing numbers remaining
print("Missing values per column after cleaning:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df, palette='Set2')
plt.title('Overall Survival Count (0 = Died, 1 = Survived)')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', hue='Sex', data=df, palette='pastel')
plt.title('Survival Rate Comparison by Gender')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', hue='Pclass', data=df, palette='magma')
plt.title('Survival Rate Comparison by Ticket Class')
plt.ylabel('Number of Passengers')
plt.show()
```
